# Исследование продаж игр различных платформ

Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). Перед вами данные до 2016 года.

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».

**Цель исследования** — проверка гипотез и выявление закономерностей, определяющих успешность игры.

**Ход исследования**

Данные из открытых источников необходимо изучить и провести обработку данных. Затем осуществить исследовательский анализ данных, провести вспомогательные для гипотез вычисления и написать вывод.

Таким образом, исследование пройдет в этап:
 1. Обзор данных.
 2. Предобработка данных.
 3. Исследовательский анализ данных.
 4. Вспомогательные вычисления, графики.
 5. Проверка гипотез.
 6. Вывод.

#### Вывод

Итак, в таблице одиннадцать столбцов.

Согласно документации к данным:

* `Name` - название игры;
* `Platform` - платформа;
* `Year_of_Release` - год выпуска;
* `Genre` - жанр игры;
* `NA_sales` - продажи в Северной Америке (миллионы проданных копий);
* `EU_sales` - продажи в Европе (миллионы проданных копий);
* `JP_sales` - продажи в Японии (миллионы проданных копий);
* `Other_sales` - продажи в других странах (миллионы проданных копий);
* `Critic_Score` - оценка критиков (максимум 100);
* `User_Score` - оценка пользователей (максимум 10);
* `Rating` - рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

В выводе первых десяти строк таблицы заметны пропущенные значения и неверные типы данных, названия столбцов не приведены к нижнему регистру. Нужно перейти к обработке данных.

### Предобработка данных

#### Изменение названий столбцов

```python
data.columns # перечень названий столбцов таблицы data
```

```python
data.columns = [column.lower() for column in data.columns] # приведение названий столбцов к нижнему регистру
```

#### Обработка пропусков

```python
data.isna().sum() # определение количества пропусков в датафрейме
```

Пропуски в столбце с названиями игр никак не получится заполнить, поэтому можно заменить их на названия-заглушки.

```python
data['name'] = data['name'].fillna('unknown')
```

Наверняка мы никак не можем заполнить год релиза игры, жанр и рейтинг ESRB без потери качества информации, поэтому лучше жанр и рейтинг оставить как "unknown", а год выпуска заменить на -1.

```python
data['year_of_release'] = data['year_of_release'].fillna(-1)
```

```python
data['genre'] = data['genre'].fillna('unknown')
```

```python
data['rating'] = data['rating'].fillna('unknown')
```

```python
data.loc[(data['name'] == 'unknown') | (data['genre'] == 'unknown')]
```

Получается, что всего две строки из всего датафрейма с неизвестными названиями игр и жанров, а еще оценками пользователей и критиков. Год выпуска аж 1993, что является точно не актуальной информацией, тогда их можно удалить.

```python
data = data.loc[(data['name'] != 'unknown') | (data['genre'] != 'unknown')]
```

Известной пользовательской оценки у игры может и не быть, поэтому заполняем неизвестные значения -1.

```python
data['user_score'] = data['user_score'].fillna(-1)
```

В столбце `user_score` также встречается значение "tbd". TBD - аббревиатура от английского To Be Determined (будет определено) или To Be Decided (будет решено). Используется, если какая-то информация еще не определена или решение по вопросу не принято. Следовательно, мы никак не можем знать эту информацию, заменяем на значения-заглушки.

```python
data.loc[data['user_score'] == 'tbd', 'user_score'] = -1
```

Заполнить пропуски в оценке критиков также невозможно: причиной пропусков может быть отсуствие оценки вообще для старых или не очень известных игр.

```python
data['critic_score'] = data['critic_score'].fillna(-1)
```

```python
data.isna().sum()
```

#### Изменение типов данных

```python
data['year_of_release'] = data['year_of_release'].astype('int')
```

```python
data['user_score'] = data['user_score'].astype('float')
```

```python
data['critic_score'] = data['critic_score'].astype('int')
```

#### Подсчет суммарных продаж в регионах

```python
data['total_sales'] = data['na_sales'] + data['eu_sales'] + data['jp_sales'] + data['other_sales']
```

#### Вывод

Предобработка обнаружила три проблемы в данных:

* пропущенные значения,
* неправильные типы данных,
* неудобные названия столбцов.

* Посмотрите, как влияют на продажи внутри одной популярной платформы отзывы пользователей и критиков. Постройте диаграмму рассеяния и посчитайте корреляцию между отзывами и продажами. Сформулируйте выводы.

```python
PS4_user_score = actual_data.loc[
    (actual_data['platform'] == 'PS4') & (actual_data['user_score'] != -1)
][['total_sales', 'user_score']]
    
PS4_user_score.plot(kind='scatter', x='user_score', y='total_sales', title='PS4', grid=True)

print('Корреляция Пирсона составляет', PS4_user_score['total_sales'].corr(PS4_user_score['user_score']))
```

Корреляция Пирсона отрицательная, но очень маленькая, близкая к нулю. Соответственно, никакой корелляции между пользовательской оценкой и продажами нет.

```python
PS4_critic_score = actual_data.loc[
    (actual_data['platform'] == 'PS4') & (actual_data['critic_score'] != -1)
][['total_sales', 'critic_score']]
    
PS4_critic_score.plot(kind='scatter', x='critic_score', y='total_sales', title='PS4', grid=True)

print('Корреляция Пирсона составляет', PS4_critic_score['total_sales'].corr(PS4_critic_score['critic_score']))
```

Корреляция Пирсона положительная, составляет примерно 0,4. Следовательно, корреляция между оценками критиков и продажами слабая.

* Соотнесите выводы с продажами игр на других платформах.

```python
def platform_sales(platform):
    user_score_data = actual_data.loc[
        (actual_data['platform'] == platform) & (actual_data['user_score'] != -1)
    ][['total_sales', 'user_score']]
    
    user_score_data.plot(kind='scatter', x='user_score', y='total_sales', title=platform, grid=True)
    
    print('Корреляция Пирсона между продажами и пользовательской оценкой для платформы ', 
          platform, 'составляет', user_score_data['total_sales'].corr(user_score_data['user_score']))
    
    critic_score_data = actual_data.loc[
        (actual_data['platform'] == platform) & (actual_data['critic_score'] != -1)
    ][['total_sales', 'critic_score']]
    
    critic_score_data.plot(kind='scatter', x='critic_score', y='total_sales', title=platform, grid=True)
    
    print('Корреляция Пирсона между продажами и оценкой критиков для платформы ', 
          platform, 'составляет', critic_score_data['total_sales'].corr(critic_score_data['critic_score']), '\n')
```

```python
for platform in actual_data['platform'].unique():
    platform_sales(platform)
```

Для других платформ ситуация остается прежней: оценки пользователей и критиков слабо коррелируют с продажами.

### Проверка гипотез

* Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

Нулевой гипотезой здесь будет предположение о том, что средние пользовательские рейтинги платформ Xbox One и PC равны, а альтернативной - разные. Проверяем гипотезу о равенстве средних двух генеральных совокупностей:

<div class="alert alert-success">
<font size="4", color= "seagreen"><b>✔️ Комментарий ревьюера</b></font>
    <br /> 
    <font size="3", color = "black">
<br /> Гипотезы сформулированы верно

```python
alpha = 0.05

results = st.ttest_ind(
    actual_data.loc[(actual_data['platform'] == 'XOne') & (actual_data['user_score'] != -1), 'user_score'], 
    actual_data.loc[(actual_data['platform'] == 'PC') & (actual_data['user_score'] != -1), 'user_score'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Следовательно, нулевую гипотезу не отвергаем. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

```python
# Приведены два датасета: сумма покупок, совершённых за месяц посетителями ...

sample_1 = [3071, 3636, 3454, 3151, 2185, 3259, 1727, 2263, 2015,
2582, 4815, 633, 3186, 887, 2028, 3589, 2564, 1422, 1785,
3180, 1770, 2716, 2546, 1848, 4644, 3134, 475, 2686,
1838, 3352]
sample_2 = [1211, 1228, 2157, 3699, 600, 1898, 1688, 1420, 5048, 3007,
509, 3777, 5583, 3949, 121, 1674, 4300, 1338, 3066,
3562, 1010, 2311, 462, 863, 2021, 528, 1849, 255,
1740, 2596]
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем гипотезу
results = st.ttest_ind(
sample_1,
sample_2)
print('p-значение:', results.pvalue)
if (results.pvalue < alpha):
    print("Отвергаем нулевую гипотезу")
else:

    print("Не получилось отвергнуть нулевую гипотезу")
```

* Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Нулевая гипотеза такова: средние пользовательские рейтинги жанров Action и Sports равные. Альтернативная же гипотеза говорит о том, что они разнятся. Проверяем гипотезу о равенстве средних двух генеральных совокупностей:

```python
alpha = 0.05

results = st.ttest_ind(
    actual_data.loc[(actual_data['genre'] == 'Action') & (actual_data['user_score'] != -1), 'user_score'], 
    actual_data.loc[(actual_data['genre'] == 'Sports') & (actual_data['user_score'] != -1), 'user_score'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Нулевую гипотезу отвергаем, p-значение меньше порогового. Тогда, согласно альтернативной гипотезе, средние пользовательские рейтинги жанров Action и Sports разные.

```python
# check
# круги + категория другие
def graph (df, year, region, name, axes):
    
    df = df.query('year_of_release >= @year')
    
    sales = df.pivot_table(index='platform', 
                           values=region, 
                           aggfunc='sum').nlargest(5, region)
    
    sales = sales.reset_index()
    
    sales = (
            sales.append({'platform': 'Other', region: df[region].sum() 
                       - sales[region].sum()}, ignore_index= True)
         )
    
    
    sales.columns = ['platform', 'sales']
      
    labels_c=sales.platform
    colours = {'Wii':'C0', 'NES':'C1', 'GB':'C2', 'DS':'C3', 'X360':'C4', 
    'PS3':'C5', 'PS2':'C6', 'SNES':'C7', 'GBA':'C8',
               'PS4':'steelblue', '3DS':'orange', 
               'N64':'C11', 'PS':'C12', 'XB':'C13', 'PC':'C14', '2600':'C15', 'PSP':'C16', 
               'XOne':'C17',
               'WiiU':'C18', 'GC':'C19', 'GEN':'C20', 'DC':'C21', 'PSV':'C22', 
               'SAT':'C23', 'SCD':'C24', 'WS':'C25', 'NG':'C26', 
               'TG16':'C27', '3DO':'C28', 'GG':'C29', 'PCFX':'C30', 'Other':'darkred'}
   
    sales.plot(kind='pie',
               y="sales",
               
               autopct='%1.0f%%',
               wedgeprops={'linewidth': 3.0, 'edgecolor': 'white'},
               textprops={'size': 'x-large'}, 
               labels= labels_c,
               colors=[colours[key] for key in labels_c],
               legend=False, 
               title = f"Популярность платформ в {name} ", 
               ax = axes).set(ylabel='')
    
    plt.tight_layout()
```
